### If-Statements w/ User Defined Functions:

```python
# Script to define price_label function
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# Applying price_label function to all rows - creating a column 'price_range'
df['price_range'] = df.apply(price_label, axis=1)
```

Error above can be ignored, its letting you know you are only affecting a slice of a larger dataframe due to df being only the first 1,000,000 rows.  Suggests using .loc to apply price_label function to entire dataframe.

Looking at the data, I believe its ideal to classify the hours that have 2,400,000 visits or above as being periods with the highest user activity.

Additionally, hours where visitors are only at 450,000 or less will be counted as the least busy periods.  This will leave anything inbetween as the "normal" periods.

```python
# Creating the results for labelling the order_hour_of_day coloumn
result3 = []
for x in ords_prods_merge['order_hour_of_day']:
    if (x == 10) or (x == 11) or (x == 14) or (x == 15) or (x == 13) or (x == 12) or (x == 16) or (x == 9):
        result3.append('Most orders')
    elif (x == 23) or (x == 6) or (x == 0) or (x == 1) or (x == 5) or (x == 2) or (x == 4) or (x == 3):
        result3.append('Fewest orders')
    else:
        result3.append('Average orders')
```

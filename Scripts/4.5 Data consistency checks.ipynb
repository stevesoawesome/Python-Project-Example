## Missing Values:

```python
# Defining path and importing data
path = r'C:\Users\steve\Documents\CF - Instacart Project\02 Data'
df_prods=pd.read_csv(os.path.join(path,'Original Data','products.csv'), index_col=False)
```

```python
# To look for nulls and add them up
df_prods.isnull().sum()
```

```python
# df_prods['product_name'].isnull()==True will create a list of all entries in the column and label it as true or false
# if it is null or not.
# df_prods[the entire thing above] creates a subset of that output, containing only values that are null
df_nan = df_prods[df_prods['product_name'].isnull() == True]
```

```python
# Table created from the above
df_nan
```

```python
# Creating new dataframe after removing null entries from product_name
df_prods_clean = df_prods[df_prods['product_name'].isnull() == False]
```

```python
# Varifying new dataframe
df_prods_clean
```

Alternate way to drop null values is to use .dropna, in this specific case you would use df_prods.dropna(subset = ['product_name'], inplace = True).

```python
# One way to view this for sake of learning format:
# var.dropna()
# var.dropna(inplace = True)
# var.dropna(subset = ['col_name'], inplace = True)
```

## Duplicates:

```python
# Command to look for dupes and create a new dataframe showing all dupes
df_dups = df_prods_clean[df_prods_clean.duplicated()]
```

A way to look at this:

make a new dataframe = createsubset of [dataframe you're using.function to find duplicates()]

```python
df_dups
```

```python
# Command to create a new table with no duplicates
df_prods_clean_no_dups = df_prods_clean.drop_duplicates()
```

```python
# Saving new csv file using cleaned up products dataframe
df_prods_clean_no_dups.to_csv(os.path.join(path,'Prepared Data','products_checked.csv'))
```
